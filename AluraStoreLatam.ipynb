### Importación de datos

# 4. Productos más y menos vendidos

```python
mas_vendido=tienda['Producto'].value_counts().idxmax()
menos_vendido=tienda['Producto'].value_counts().idxmin()
print(f'el producto mas vendido es {mas_vendido} y el menos vendido {menos_vendido}')
vendidos=tienda['Producto'].value_counts()
vendidos.plot(kind='bar',figsize=(12,3),color='purple',xlabel='Producto',ylabel='Cantidad')

```

#Visualizaciones


```python

tienda['Fecha de Compra'] = pd.to_datetime(tienda['Fecha de Compra'])
tienda['Mes'] = tienda['Fecha de Compra'].dt.to_period('M')
ventas_mensuales = tienda.groupby('Mes')['Precio'].mean()
ventas_mensuales.plot(kind='line', figsize=(20,4),color='purple',xlabel='Año-Mes', ylabel='Ventas')
```

```python
tienda2['Fecha de Compra'] = pd.to_datetime(tienda2['Fecha de Compra'],dayfirst=True)
tienda2['Mes'] = tienda2['Fecha de Compra'].dt.to_period('M')
ventas_mensuales2 = tienda2.groupby('Mes')['Precio'].mean()
ventas_mensuales2.plot(kind='line', figsize=(20,4),color='green',xlabel='Año-Mes', ylabel='Ventas')
```

```python
tienda3['Fecha de Compra'] = pd.to_datetime(tienda3['Fecha de Compra'])
tienda3['Mes'] = tienda3['Fecha de Compra'].dt.to_period('M')
ventas_mensuales3 = tienda3.groupby('Mes')['Precio'].mean()
ventas_mensuales3.plot(kind='line', figsize=(20,4),color='blue',xlabel='Año-Mes', ylabel='Ventas')
```

```python
tienda4['Fecha de Compra'] = pd.to_datetime(tienda4['Fecha de Compra'],dayfirst=True)
tienda4['Mes'] = tienda4['Fecha de Compra'].dt.to_period('M')
ventas_mensuales4 = tienda4.groupby('Mes')['Precio'].mean()
ventas_mensuales4.plot(kind='line', figsize=(20,4),color='yellow',xlabel='Año-Mes', ylabel='Ventas')
```
